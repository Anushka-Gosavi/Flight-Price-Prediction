```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
 # use of %matplotlib inline --> we can show plots in this model without using .show()
```

```python
train_df=pd.read_excel('/content/Data_Train.xlsx')
test_df=pd.read_excel('/content/Test_set.xlsx')
# test dataset dont have price column. as weare making this model only to predict price
```

```python
train_df.head()
```

```python
# S1 : Combine both Dataset
final_df=pd.concat([train_df,test_df])
```

```python
final_df.info()
```

S2 : Our Aim is to convert all objects into int (Categorical features --> Numerical Features)

```python
# Feature Engineering on Date_of_Journey
#creating Date, Month and Year columns
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]


                       #or Use Lambda Function
 #final_df['Date']=final_df['Date_of_Journey'].apply(lambda x:x.split('/')[0])
```

```python
final_df['Date']=final_df['Date'].astype(int)
final_df['Month']=final_df['Month'].astype(int)
final_df['Year']=final_df['Year'].astype(int)
```

```python
final_df=final_df.drop('Date_of_Journey',axis=1)
```

```python
#feature engineering on Arrival_Time and Dep_Time
final_df['Arrival_Time']=final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df['Arrival_Hour']=final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_Minute']=final_df['Arrival_Time'].str.split(':').str[1]
```

```python
#still in object type, convert into int
final_df['Arrival_Hour']=final_df['Arrival_Hour'].astype(int)
final_df['Arrival_Minute']=final_df['Arrival_Minute'].astype(int)
```

```python
final_df.info()
```

```python
final_df=final_df.drop('Arrival_Time',axis=1)
```

```python
final_df.isnull().sum()
```

```python
#feature engineering on departure_time

final_df['Dep_hr']=final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_min']=final_df['Dep_Time'].str.split(':').str[1]
```

```python
final_df['Dep_hr']=final_df['Dep_hr'].astype(int)
final_df['Dep_min']=final_df['Dep_min'].astype(int)
```

```python
final_df=final_df.drop(['Dep_Time'], axis=1)
```

```python
final_df.info()
```

```python
# FE on Total Stops
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'NaN':1})
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df=final_df.drop('Route',axis=1)
```

```python
final_df.head()
```

```python
# FE on Duration
final_df['Duration_hr']=final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
final_df[final_df['Duration_hr']=='5m']
```

```python
# drop abouve 2 records
final_df.drop(6474,axis=0,inplace=True)
final_df.drop(2660,axis=0,inplace=True)
```

```python
final_df['Duration_hr']=final_df['Duration_hr'].astype(int)
```

```python
final_df.drop(columns='Duration',inplace=True)
```

```python
#FE on Airline
final_df['Airline'].unique()
```

```python
# we will do label encoding for this
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
final_df['Airline']=le.fit_transform(final_df['Airline'])
```

```python
# same for Source, Destination and Additional Info
final_df['Source']=le.fit_transform(final_df['Source'])
final_df['Destination']=le.fit_transform(final_df['Destination'])
final_df['Additional_Info']=le.fit_transform(final_df['Additional_Info'])
```

```python
# one can do similar thig using OneHotEncoding
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
final_df['Airline']=ohe.fit_transform(final_df['Airline'].values.reshape(-1,1))
#This code reshapes the final_df['Airline'] column into a 2-dimensional array using the reshape method and then applies the fit_transform method of the OneHotEncoder to the reshaped array.

```
